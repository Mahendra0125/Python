```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#Resizing the graph
plt.figure(figsize=(5,3),dpi=100)

#Plotting a graph with label, markersize, color
plt.plot(x,y, 'r.-',label = '2x', markersize=12)

plt.title('Our First Graph!', fontdict ={'fontname':'Comic Sans MS', 'fontsize':15})

#Using Numpy to get a plot
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2,'b^-',label = 'x^2')
plt.plot(x2[5:], x2[5:]**2,'b^--')

#Labelling the axis
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

#Changing the x and y ticks
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

#Saving the figure
plt.savefig('My Graph.png',dpi = 300)

#Adding a Legend
plt.legend()

plt.show()
```
